```python
#Create feature matrix without relapse column
```

```python
X = finaldf[finaldf.columns.values]
X.drop('Relapse', axis = 1, inplace = True)

y = finaldf.Relapse
```

```python
X
```

```python
## Customary Decision Tree
```

```python
y
```

```python
train_X, test_X, train_y, test_y = cross_validation.train_test_split(X, y, train_size = 0.6, random_state = 0)
```

```python
model = tree.DecisionTreeRegressor(random_state = 0).\
    fit(train_X, train_y)
```

```python
train_y_hat = model.predict(train_X)
print np.sqrt(metrics.mean_squared_error(train_y, train_y_hat))

test_y_hat = model.predict(test_X)
print np.sqrt(metrics.mean_squared_error(test_y, test_y_hat))
```

```python
modelDT = tree.DecisionTreeRegressor(max_depth = 4, random_state = 0).\
    fit(train_X, train_y)
```

```python
train_y_hat = modelDT.predict(train_X)
print np.sqrt(metrics.mean_squared_error(train_y, train_y_hat))

test_y_hat = modelDT.predict(test_X)
print np.sqrt(metrics.mean_squared_error(test_y, test_y_hat))
```

```python
dot_data = externals.six.StringIO()
tree.export_graphviz(modelDT, out_file = dot_data, feature_names = X.columns)
%dotstr dot_data.getvalue()
```

```python
## Feature Importance
```

```python
sorted(zip(modelDT.feature_importances_, X.columns.values), reverse = True)
```

```python
cross_val_score(modelDT, X, y, cv=10)
```

```python
parameters = {'max_depth':range(3,20)}
clf = grid_search.GridSearchCV(tree.DecisionTreeClassifier(), parameters, n_jobs=4)
```

```python
## Grid Search takes time. 
```

```python
clf.fit(X=X, y=y)
tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_) 
```

```python
# Lets try with max features as 100 and then 1000. 
```

```python
## Random Forest. With 1000 max features and 10000 estimator. Make it odd to break ties
```

```python
modelRF = ensemble.RandomForestClassifier(n_estimators = 10001,
        max_features = 1000,
        min_samples_leaf = 5,
        oob_score = True,
        random_state = 0).\
    fit(train_X, train_y)
```

```python
modelRF.oob_score_
```

```python
# Feture Importance
```

```python
sorted(zip(modelRF.feature_importances_, X.columns.values), reverse = True)
```

```python
# Try with no max features. 
```

```python
modelRFNoMaxFeatures = ensemble.RandomForestClassifier(n_estimators = 10001,
        oob_score = True,
        random_state = 0).\
    fit(train_X, train_y)
```

```python
sorted(zip(modelRFNoMaxFeatures.feature_importances_, X.columns.values), reverse = True)
```

```python
modelRFNoMaxFeatures.oob_score_
```

```python
##Cross Validation
```
